```python
# Normalize column names
df.columns = df.columns.str.strip().str.lower()

# Confirm new column names
print(df.columns.tolist())

# Parse datetime correctly
df['date rptd'] = pd.to_datetime(df['date rptd'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
df['date occ'] = pd.to_datetime(df['date occ'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

# Extract hour
df['hour'] = df['date occ'].dt.hour

# Check hourly distribution
print(df['hour'].value_counts().sort_index())
```

```python
#Checking whats wrong with the DATE Rptd
import re

# Now you can use:
unique = df['date rptd'].astype(str).unique()
bad = [val for val in unique if not re.match(r"\d{2}/\d{2}/\d{4}$", val)]
print(bad[:20])  # prints up to 20 invalid entries

unique = df['date rptd'].astype(str).unique()
print([val for val in unique if not re.match(r"\d{2}/\d{2}/\d{4}$", val)])
```

```python
#Next steps after importing
import re

unique = df['date rptd'].astype(str).unique()
bad = [val for val in unique if not re.match(r"\d{2}/\d{2}/\d{4}$", val)]
print("Bad examples:", bad[:20])

# Optional: convert using flexible parsing
df['date rptd'] = pd.to_datetime(df['date rptd'], infer_datetime_format=True, errors='coerce')
print("Missing after parsing:", df['date rptd'].isna().sum())
```

```python
#Fixing Date Rptd parsing
df['date rptd'] = pd.to_datetime(
    df['date rptd'], 
    errors='coerce'  # invalid entries become NaT
)
```

# Feature Understanding

```python
df['month'].value_counts().sort_index().plot()
```

```python
df['year'].value_counts().sort_index().plot(kind = 'bar')
```

```python
df = df[~((df['month'] ==12) & (df['year'] ==2025))]
```

```python
df.sort_values(by = 'date occ')
```

```python
crime_over_time = df.groupby(['year','month']).size().plot()

crime_over_time
```

```python
plt.hist(df_age['vict age'], edgecolor = 'black', bins = 15)
```

```python
df['DATE OCC no time'] = df['date occ'].dt.date
```

```python
crime_count_per_day = df.groupby(['DATE OCC no time', 'crm cd desc']).size()
```

```python
top_20_crimes = crime_count_per_day.groupby('crm cd desc').mean()\
                    .reset_index(name = 'average_daily_count').sort_values(by = 'average_daily_count', ascending = False).head(20)
```

```python
top_20_crimes_sorted = top_20_crimes.sort_values(by = 'average_daily_count', ascending = True)
```

```python
plt.barh(top_20_crimes_sorted['crm cd desc'], top_20_crimes_sorted['average_daily_count'])
```

```python
plt.figure(figsize = (12,8))


#Plot 1 - Line Chart
plt.subplot(2,2,1)
crime_plot = df.groupby(['year','month']).size().plot(color = 'green', linewidth = 2)
crime_plot.set_xlabel('Year - Month', fontsize = 10)
crime_plot.set_title('Crime over Time')

#Plot 2 - Histogram
plt.subplot(2,2,2)
plt.hist(df_age['vict age'], edgecolor = 'black', bins = 15)
plt.tight_layout()
plt.xlabel('Age', fontsize = 10)
plt.title('Distribution of Victim Age')

#Plot 3 - Horizontal bar
plt.figure()
plt.barh(top_20_crimes_sorted['crm cd desc'], top_20_crimes_sorted['average_daily_count'], color = 'orange')
plt.title('Top 20 Crimes')
```
